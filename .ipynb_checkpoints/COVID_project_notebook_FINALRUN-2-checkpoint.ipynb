## 0.1 Cluster setup
***

Since the group members are evenly distributed between Emilia-Romagna and Veneto, at first we were looking for a method to link together our machines through the net.

We succeded in the cluster creation using `ngrok`, a service that allows to expose any networked service that runs over TCP tunnels.
In its free version it allows to create 4 TCP tunnels for each machine with which we can establish a connenction between dask scheduler and the workers.
The idea was to dedicate one machine to the the scheduler, the dashboard, a worker and its nanny port, and all the others to two workers and their respective nanny ports.

The cluster was functioning, it could properly execute some toy-examples and the country search task.
For more intensive tasks like the word counting one we faced some `CommClose` errors for which the computation could not converge.

## 1.1 Word counter distributed algorithm
***

In NLP the frequency of appearance of a word in the documents is a very important quantity, so we want to compute it in the whole dataset of papers we are studying.

To complete this task we adopted a **Map-Reduce** approach to get these quantities, where a first count of how many times each of the words appear in each of the papers is performed (**Map** phase) and then the number of appearancies of the words between the documents is summed altogether (**Reduce** phase).

After the `body_text` field is plucked from the list of files, the **map** phase is handled by the function `word_counter` that accomplishes three main tasks:

* concatenating the `text` field for each of the sections of the `json` papers;

* cleaning the resulting string, by deleting unwanted characters and numbers, and removing *stopwords* (common words in the whole english language not really interesting to process);

* counting the occurencies of each word in the text.






````python
def word_counter(body_text_list):

    '''Function to concatenate the strings of text from the json, clean them from unwanted characters and     '''
        
    #concatenation of every section into one text file
    total_text = ''
    for text_section in body_text_list:
        total_text += text_section['text']
        
    total_text = total_text.lower()#lowercase everything
        
    # deteleting all unwanted characters
    
    forbidden_chars = ".,()[]:*><;%/+-=1234567890"
    
    for forbidden_ch in forbidden_chars:
        
        total_text = total_text.replace(forbidden_ch,"")
    
    total_text = total_text.replace('"','')
        
        
    text_list = total_text.split() # transform into a list
    
    stopw = stopwords.words('english') 
    
    alphabet = [i for i in'abcdefghilmnopqrstuvz']
    otherw = stopw + ['fig', 'et', 'al', 'also', 'may', 'using', 'use', 'used', 'one', 'two', 'however', 'can', 'could'] + [str(i) for i in range(11)]  + alphabet 
    text_list = [i for i in text_list if i not in otherw]    
    
    words, counts = np.unique(text_list, return_counts  = True) # count the words
    
    word_counts = [(key , count) for key,count in zip(words, counts)]

    return word_counts
````

For the **reduce** phase we used the *foldby* functionality of dask bags, where a certain operation is applied to elements of the bag that share a certain key.

 The bags need to be flattened in order to have the elements of the bag being the signle tuple of word + number of occurency in paper. The operation applied is the sum of the occurencies.

 The result of this operation is a list of bags containing word + occurency in the whole corpus of papers. Below it is possible to find the code to run the proces described above.

````python
def add2(x1,x2):
    return x1 + x2[1]

def add3(a,b):
    return a+b
````

````python
b_conc = b.pluck('body_text').map(word_counter).flatten().foldby(lambda x:x[0],binop=add2,initial=0,combine=add3,combine_initial=(0))
word_count = b_conc.compute()
````

````python
word_count = {key[0]:key[1][0] for key in word_count}

#sorting the dictionary
word_sort = dict(sorted(word_count.items(), key=lambda item: item[1])[::-1])
````

## 1.2 Test with different workers and partitions.
***

We changed the number of workers and partitions to see how the performances are affected. The partitions chosen are always a multiple of the number of workers so that the partitions are split evenly among them. We chose $1,4,8,32,64,128$ as the number of partitions and $1,2,4,8$ as the number of workers. Each worker has 1 thread.

The results are shown in a seaborn heatmap.

````python
partitions = [1,4,8,32,64,128]
workers = [1,2,4,8]
## ----------------------------------------------------------------------------------------
##   so the worker should be initialized to have as resources "ix" respectively: 4,3,2,2,1,1,1,1
## ----------------------------------------------------------------------------------------
times_array=np.zeros([len(workers),len(partitions)])

#b = db.read_text(flistgit).map(json.loads).repartition(partitions[j])

loading = db.read_text(flistgit).map(json.loads).repartition(8).compute()

for i in range(len(workers)):
    for j in range(len(partitions)):
        
        with dask.annotate(resources={'ix': len(workers)-i}):
            bag = db.from_sequence(loading)
            #b = db.read_text(flistgit).map(json.loads).repartition(partitions[j])
            time1 = time.time()
            word_count = bag.repartition(partitions[j]).pluck('body_text').map(word_counter).flatten().foldby(lambda x:x[0],binop=add2,initial=0,combine=add3,combine_initial=(0,0)).compute()
            times_array[i,j] = time.time() - time1
````

````python
import seaborn as sns

df_times=pd.DataFrame(times_array,workers,partitions)

fig,ax=plt.subplots(figsize=[10,10])

sns.heatmap(df_times,cmap='inferno_r',square=True,annot=True,fmt='.3f',cbar=True,ax=ax, cbar_kws={"shrink": 0.56})
plt.title('Computing time as a function of number of workers and partitions', fontweight='bold')
plt.ylabel('Workers')
plt.xlabel('Partitions')
plt.show()
````

As we can see, increasing the number of workers usually makes the process run faster. If the number of partitions is too high, the process becomes rather slow because of the time needed to merge all of the results. The best result has been obtained with 8 workers (8 threads) and 8 partitions: we can conclude that in this special case the optimal situation is reached when the number of partitions is the same or a small multiple of the number of workers or threads.

## 3.1.1 Get the embedding for the title of the papers
***


In NLP a common technique to perform analysis over a set of texts is to transform the text to a set of vectors each one representing a word inside a document. At the end of the pre-processing the document will be transformed into a list of vectors or a matrix of $n \times m$ where $n$ is the number of words in the document and $m$ is the size of the vector that represents the word $n$. This technique is called **Embedding**.

In this section, a transformation of the titles of all papers into their embedding version has been performed exploiting a pre-trained model.
The pre-trained model we used is `wiki.en.vec` developed by Bojanowski et al.(2016) and it is based on the Skip Gram technique. The results are stored in a dictionary in form $key:vector$, where the vectors are each one of dimension 300.

````python
def load_wordlist(fname):
    words = {}
    
    with io.open(fname, 'r', encoding='utf-8', newline='\n', errors='ignore') as fin:
        #n, d = map(int, fin.readline().split())
        # n = number of words
        # d = vector dimension
        
        
        #embeddings = np.zeros((n, d), dtype=np.float32)
        
        for i, line in enumerate(fin):
            tokens = line.rstrip().split(' ')
            words[tokens[0]] = i
            #embeddings[i, :] = np.array(tokens[1:], dtype=np.float32) 
                
        return words #, embeddings
````

Since the dictionary with the words and their embeddings is really large (6 GB), we chose to read the lines of the file one by one just when necessary.
In order to do this, we needed a dictionary containing the information about which line of the file every word could be found. This is defined in the `words` dictionary that is computed down below and that is saved to a file, so that in the future it is faster to to read. This dictionary contains the words and their line index.

````python
words =  json.load(open('words_finder.json'))
````

````python
def sanitize_text(text):
    '''Remove unwanted characters and transform the text to lowercase'''
    forbidden_chars = ".,()[]:*><;%/+-=1234567890"
    for forbidden_ch in forbidden_chars :
        text = text.replace(forbidden_ch,"")
    return(text.lower())

def extract_title(b):
    '''Get paper id and title words list from a paper ''''
    title = b['metadata']['title']
    title = sanitize_text(title)
    title = title.split()
    return {'paper_id':b['paper_id'], 'title': title}

def get_ind_list(b):
    '''Get indices of the of each word in the title so that it can be more easily be read from from the embedding file'''

    # words = parole con indici
    title = b['title']
    ind = []
    for word in title:
        ind.append(words.get(word))
    return {'paper_id':b['paper_id'], 'title': ind}

def get_embeddings_list(b):

    '''Get embeddings of titles if indices list is given in input'''
    
    inds = b['title']
    
    fp = io.open('wiki.en.vec', 'r', encoding='utf-8', newline='\n', errors='ignore')
    
    emb = []
    
    inds = list(filter(None, inds))
    
    if len(inds)==0:
        return []

    m = max(inds)

    for i, line in enumerate(fp):
        if i in inds:
            tokens = line.rstrip().split(' ')
            emb.append(np.array(tokens[1:], dtype=np.float32))
            
        elif i > m:
            break
    fp.close()
    return {'paper_id':b['paper_id'], 'title_embedding': emb}
````

````python
intermediate_results = b.map(extract_title).map(get_ind_list).persist()#.map(get_embeddings_list).persist()
````

````python
results = intermediate_results.map(get_embeddings_list).compute()
````

````python
title_embeddings_to_file = []

for title in results: 
    try:
        title_embeddings_to_file.append({'paper_id':title['paper_id'],'title_embedding': [a.tolist() for a in title['title_embedding']]})
    except:
        pass
    
with open('title_embeddings.json','w') as f:

    json.dump(title_embeddings_to_file,f)
````
